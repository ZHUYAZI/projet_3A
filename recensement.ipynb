```python
with open("D:\\projet_3A\\tisseo_gtfs\\stops.txt") as f:
    next(f)
    for line in f:
        # assuming tabs and not multiple space characters
        stop_id,stop_code,stop_name,stop_lat,stop_lon,location_type,parent_station,wheelchair_boarding = line.strip().split(",") 
        data = {
            "stop_id": stop_id,
            "stop_code":stop_code,
            "stop_name":stop_name,
              "location" : {
                    "type" : "Point",
                    "coordinates" : [
                    float(stop_lon),float(stop_lat)
                    ]
            }
        }
        stops.insert_one(data).inserted_id
        
```

```python
for i in a13.find():
    zone = i['fields']['geo_shape']
    
    query=db.stops.find({
           "location": {
           "$geoIntersects": {
           "$geometry": zone
         }
       }
     })
    
    stop_number = len(list(query))
    print(i['_id'],stop_number)
    db.a13.update_one(
       {"_id" : i['_id']},
       {
           "$set":{ 'num_stop': int(stop_number)}
       }
    )
```
