```python
import pymongo
from pymongo import MongoClient
import pprint
import pandas as pd
```

```python
client = MongoClient()
client = MongoClient('localhost', 27017)
client.list_database_names()
```

```python
db = client['recensement']
db.list_collection_names()
```

```python
terrase=db['terrase']
a13 = db['a13']
p13=db['p13']
```

```python
terrase.count_documents({})
```

```python
for i in terrase.find():
    pprint.pprint(i)
```

```python
codination=[]
for i in a13.find():
    codination.append(i['fields']['geo_shape'])
    
codination
```

```python
nombre_terrase=[]
for location in codination:
    query=db.terrase.find({
           "geometry":{
               "$geoIntersects": {
            "$geometry": location
           }
       }
     })
    nombre_terrase.append(len(list(query)))
```

```python
nombre_terrase
```

```python
sum(nombre_terrase)
```

```python
nombre_stop=[]
for i in a13.find():
    nombre_stop.append(i['num_stop'])
nombre_stop
```

```python
nombre_logement=[]
for i in a13.find():
    nombre_logement.append(i['fields']['p13_log'])
nombre_logement
```

```python
name_region=[]
for i in a13.find():
    name_region.append(i['fields']['libelle_du_grand_quartier'])
name_region    
```

```python
from area import area
```

```python
surface_region=[]
for i in a13.find():
    surface_region.append(area(i['fields']['geo_shape']))
surface_region
```

```python
from geopy.distance import great_circle
```

```python
centre_capitol = (43.604401,1.443417)
distance=[]
for i in a13.find():
    distance.append(great_circle(centre_capitol,i['fields']['geo_point']).km)

distance
```

```python
name2_region=[]
population=[]
for i in p13.find():
    name2_region.append(i['fields']['libelle_des_grands_quartiers'])
    population.append(i['fields']["p13_pop"])
population
```

```python
name2_region
```

```python
df1 = pd.DataFrame(list(zip(name_region,nombre_terrase, nombre_logement,surface_region,distance)),
                   columns =['name_region','nombre_terrase', 'nombre_logement','surface_region','distance'])
df1
```

```python
df2 = pd.DataFrame(list(zip(name2_region, population)), 
               columns =['name_region', 'population']) 
df2
```

```python
dataset=pd.merge(df1, df2, on="name_region")
dataset
```

```python
dataset.to_csv("population_dataset.csv")
```
